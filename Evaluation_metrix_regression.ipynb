```python
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Example true and predicted values
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# Mean Squared Error (MSE)
mse = mean_squared_error(y_true, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_true, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# R-squared (R²)
r2 = r2_score(y_true, y_pred)
print(f"R-squared (R²): {r2:.2f}")
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = np.random.randn(100,5)
```

```python
X.shape
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k,random_state=0,init='k-means++')
  kmeans.fit(X_pca)
  wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,4))
plt.plot(range(1,11),wcss, marker='o')
plt.grid()
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k,init='k-means++',random_state=0)
kmeans.fit(X_pca)
```

```python
labels = kmeans.labels_
```

```python
plt.scatter(X_pca[:,0],X_pca[:,1],c= labels)
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('/content/income.csv')
data
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns

# Assuming you have a DataFrame named 'data'

# Scale the 'Age' and 'Income($)' columns using Min-Max scaling
s = MinMaxScaler()
data['Age'] = s.fit_transform(data[['Age']])
data['Income($)'] = s.fit_transform(data[['Income($)']])

wcss = np.array([])
for k in range(1, len(data['Age'])):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data[['Age', 'Income($)']])  # Fit KMeans to the scaled data
    wcss = np.append(wcss, kmeans.inertia_)

wcss_dict = {x: y for x, y in enumerate(wcss)}

# Create a scatter plot for WCSS
sns.scatterplot(x=list(wcss_dict.keys()), y=list(wcss_dict.values()))
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()

# Based on the elbow method, choose an optimal number of clusters (e.g., k=3)
optimal_k = 3

# Fit KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(data[['Age', 'Income($)']])

# Add cluster labels to the DataFrame
data['clusters'] = kmeans.labels_

# Create a scatter plot of cluster centers and data points
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], label='Cluster Centers', marker='X', s=100)
sns.scatterplot(x='Age', y='Income($)', data=data, hue='clusters', palette='Set1')
plt.xlabel('Age (Scaled)')
plt.ylabel('Income ($) (Scaled)')
plt.title(f'K-means Clustering (k={optimal_k})')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(data[['Age', 'Income($)']], kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg}')
```
